```python
def get_key_and_value_counts(value_counts: pd.core.series.Series):
    result = {}
    for k,v in value_counts.iteritems():
        result[k] = v
    return result


def get_key_and_value_counts_percent(value_counts: pd.core.series.Series, default_count=5):
    result = {}
    for k,v in value_counts.iteritems():
        result[k] = v / default_count
    return result
```

```python
print("Total User:", df.shape[0])
print("views:", get_key_and_value_counts(views_counts))
print("creation date:", get_key_and_value_counts(creation_date_counts))
print("rating:", get_key_and_value_counts(rating_counts))
print("similarity score:", get_key_and_value_counts(similarity_score_counts))
print("bookmark:", get_key_and_value_counts(bookmark_counts))

data = {
    "views": get_key_and_value_counts(views_counts),
    "creation_date": get_key_and_value_counts(creation_date_counts),
    "rating": get_key_and_value_counts(rating_counts),
    "similarity_score": get_key_and_value_counts(similarity_score_counts),
    "bookmark_count": get_key_and_value_counts(bookmark_counts),
}
```

# Calculating the percentages for each question and order them

```python
def order_factors(data: dict, nbr_participant=0):
    result = {}
    for key, stats in data.items():
        factor = {}
        # factor = [{k: v} for k, v in stats.items()]
        for k, v in stats.items():
            factor[k] = (v / nbr_participant) # * 100

        values = [v for k, v in factor.items()]
        avr_sum = sum(values) / len(values)
        result[key] = avr_sum

        print(values)
        print(avr_sum)
        print()

    return result

order_factors(data=data, nbr_participant=4)
```
